# Grizzly demo

```python
import grizzly as gz
```

```python
graph = gz.read_csv('nalt.csv')
graph
```

```python
graph.decode()
```

```python
labels = graph.query('''
select *
where {
    ?concept a skos:Concept .
    ?concept ?relation ?label .
    ?label a skosxl:Label .
    ?label skosxl:literalForm ?labelstring .
}
''')
labels.decode()
```

```python
import koalas as kl
```

```python
normalized = (kl.WordFrame(labels.decode())
                .str.lower(on='labelstring', to='normalized')
                .remove_qualifiers()
                .lemmatize()
                .reset_index()
                .assign(prefix='http://test/')
                .assign(newURI=lambda x: x['prefix'] + x['index'].astype(str))
             )
normalized
```

```python
table = gz.Table.from_df(normalized)
table
```

```python
table.decode()
```

```python
newtriples = table.construct('''
?concept skosxl:altLabel ?newURI.
?newURI skosxl:literalForm ?normalized.
''')
newtriples.decode()
```

```python
newtriples.to_turtle('newtriples.ttl')
```
